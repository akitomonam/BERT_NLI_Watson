# モデルの作成

```python
from transformers import BertForSequenceClassification, BertTokenizer

sc_model = BertForSequenceClassification.from_pretrained('bert-base-uncased',  num_labels=3)
sc_model.cuda()
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
from datasets import load_dataset

def tokenize(batch):
    texts = [text1 + "[SEP]" + text2 for text1, text2 in zip(batch['text1'],batch['text2'])]
    return tokenizer(texts, padding=True, truncation=True, max_length=256)


train_data = load_dataset("csv", data_files=nli_path+"nli_train.csv", column_names=["text1","text2", "label"], split="train")
display(train_data)
train_data = train_data.map(tokenize, batched=True, batch_size=len(train_data))
display(train_data)
train_data.set_format("torch", columns=["input_ids", "label"])
display(train_data)

test_data = load_dataset("csv", data_files=nli_path+"nli_test.csv", column_names=["text1", "text2", "label"], split="train")
test_data = test_data.map(tokenize, batched=True, batch_size=len(test_data))
test_data.set_format("torch", columns=["input_ids", "label"])
```

```python
from sklearn.metrics import accuracy_score

def compute_metrics(result):
    labels = result.label_ids
    preds = result.predictions.argmax(-1)
    acc = accuracy_score(labels,preds)
    return {
        "accuracy":acc,
    }
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./",
    num_train_epochs = 4,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=32,
    warmup_steps = 500, #学習係数が0からこのステップで上昇
    weight_decay = 0.01, #重みの減衰率
    logging_dir = "./logs",
    # evaluate_during_training=True, # バグ発生、下で代替
    evaluation_strategy='steps',
    report_to=None,
)

trainer = Trainer(
    model = sc_model,
    args = training_args,
    compute_metrics = compute_metrics,
    train_dataset = train_data,
    eval_dataset = test_data,
)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"
```

```python
trainer.train()
```
